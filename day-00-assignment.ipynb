## Задание 5

Объедини таблицы из разных предметных областей в одну итоговую таблицу. В этом тебе пригодится функция [pd.merge()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html).

После каждого объединения таблиц удаляй колонку `client_id` с помощью метода [.drop()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html)

Выведи, сколько **строк и столбцов** получилось в итоговой таблице.

```python
# Объединяем таблицы appeals и client
merged_df1 = pd.merge(clients, appeals, left_on="id", right_on="client_id")
merged_df1.drop(columns='client_id', inplace=True)  # Удаляем колонку client_id

# Объединяем результаты с таблицей failures
merged_df2 = pd.merge(merged_df1, failures, left_on="id", right_on="client_id")
merged_df2.drop(columns='client_id', inplace=True)  # Удаляем колонку client_id

# Объединяем с таблицей geography
merged_df3 = pd.merge(merged_df2, geography, left_on="id", right_on="client_id")
merged_df3.drop(columns='client_id', inplace=True)  # Удаляем колонку client_id

# Объединяем с таблицей iptv
merged_df4 = pd.merge(merged_df3, iptv, left_on="id", right_on="client_id")
merged_df4.drop(columns='client_id', inplace=True)  # Удаляем колонку client_id

# Объединяем с таблицей services
merged_df5 = pd.merge(merged_df4, services, left_on="id", right_on="client_id")
merged_df5.drop(columns='client_id', inplace=True)  # Удаляем колонку client_id

# Выводим размер итоговой таблицы
print(f"Количество строк: {merged_df5.shape[0]}, Количество столбцов: {merged_df5.shape[1]}")
```
